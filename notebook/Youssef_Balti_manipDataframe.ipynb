```python
import pandas as pd
# préciser le chemin du fichier à lire
cheminDuFichierCSV = "./../data/dictionary/1757_Buchanan-J.csv" 
# demander à pandas de lire le fichier et garder ce tableau accessible dans une variable
dfDicoBuch = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")
# afficher à l'écran un échantillon du tableau
dfDicoBuch.sample(n=5)
```

### Lecture du fichier : Buchanan_PronChar_counts

```python
# préciser le chemin du fichier à lire
cheminDuFichierCSV = "./../data/dictionary/Buchanan_PronChar_counts.csv" 
# demander à pandas de lire le fichier et garder ce tableau accessible dans une variable
Buchanan_count = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")
# afficher à l'écran un échantillon du tableau
Buchanan_count.sample(n=5)
#Remplacer les Nan par des '' afin de faire la correction des noms.
Buchanan_count = Buchanan_count.fillna('')
```

```python
word_list=[]
i=0
from tqdm import tqdm
for word in tqdm(dfDicoBuch['pron']):
    string_list = list(word)
    for j in range(len(string_list)):
        for i in range(len(Buchanan_count)):
            string_list[j]=string_list[j].replace(str(Buchanan_count.iloc[i,0]),str(Buchanan_count.iloc[i,2]))
        new_string = "".join(string_list)
    word_list.append(new_string)
```

```python
#Ajouter la liste qui contient les mots vedettes au datframe existant pour l'enrichir
dfDicoBuch['motVedette'] = word_list
```

```python
# définir une fonction qui retourne quelque chose
def sizePron(strPron):
    #retourner le nombre de caractère de la prononciation
    return len(strPron)
```

```python
# appeler cette fonction sur toutes les lignes de la matrice pour remplir une nouvelle colonne sizePron grace à apply
dfDicoBuch['sizePron'] = dfDicoBuch.apply(lambda x: sizePron(x['pron']),1)
```

```python
#Affichage de 5 lignes du dataframe dfDicoBuch pour assurer qu'on a devellopé le resultat attendu
dfDicoBuch.sample(n=5)
```

```python
#Exporter le nouveau dataframe enrichie à un fichier csv
dfDicoBuch.to_csv('./../data/dfDicoBuch_Enrichi.csv',index=False,encoding='utf-8')
```

### Lecture du fichier : BigiRichSample

```python
# préciser le chemin du fichier à lire
cheminDuFichierCSV = "./../data/dataframe/BigiRich.csv" 
# demander à pandas de lire le fichier et garder ce tableau accessible dans une variable
BigiRich = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")
# afficher à l'écran un échantillon du tableau
BigiRich.sample(n=5)
len(BigiRich)
```

```python
# préciser le chemin du fichier à lire
cheminDuFichierCSV = "./../data/dfDicoBuch_Enrichi.csv" 
# demander à pandas de lire le fichier et garder ce tableau accessible dans une variable
dfDicoBuch = pd.read_csv(cheminDuFichierCSV, delimiter=",", encoding="utf8")
```

```python
#une colonne Sampa et une colonne Ipa à partir des mots vedettes obtenus (transformation/croisement de données) en utilisant ./../data/dataframe/BigiRichSample.csv
dfDicoBuch_IpaSampa=pd.merge(
    dfDicoBuch,
    BigiRich,
    how="left",
    left_on=['motVedette'],
    right_on=['hwd'])
```

##### Obtenir l'information complémentaire POS (part of speech) ( Bonus )

```python
def POS(row):
   if row['POS'] == "n." :
      return "noun"
   if row['POS'] == "a." :
      return 'adjective'
   if row['POS'] == "v." :
      return 'verb'
   if row['POS'] == "p.":
      return 'pronoun'

dfDicoBuch_IpaSampa['PartOfSpeech'] = dfDicoBuch_IpaSampa.apply (lambda row: POS(row), axis=1)
```

```python
# On affiche le nombe de lignes où la valeur de ipa n'est pas null.
selected_rows = dfDicoBuch_IpaSampa[~dfDicoBuch_IpaSampa['pronSampa'].isnull()]
len(selected_rows)
```

```python
#isoler les mots de Buchanan qui n'existent pas dans Bigi.  Tenter de comprendre les différentes causes et tenter d'expliquer comment résoudre les différents problèmes (bonus)
non_selected_rows = dfDicoBuch_IpaSampa[dfDicoBuch_IpaSampa['ipa'].isnull()]
len(non_selected_rows)
```

```python
# sauvegarder la nouvelle matrice obtenue dans ./../data/dataframe/BuchananRich.csv
dfDicoBuch_IpaSampa.to_csv('./../data/dataframe/BuchananRich.csv',index=False,encoding='utf-8')
```
